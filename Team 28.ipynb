```python
import nltk
import random
import time
import pandas as pd
import xlrd

ti=[]
u=[]
v=[]
cc=0
score=0
noun=['NN','NNS','NNP','NNPS','PDT']
pronoun=['PRP','PRP$','WP']
verb=['VB','VBG','VBD','VBN','VBZ','MD','TO']
adverb=['RB','RBR','RBS','WRP','FW']
adjective=['JJ','JJR','JJS']
preposition=['POS']
conjunction=['CC','IN']
interjuction=['UH']
article=['DT','WDT']

# def time_convert(sec):
#   mins = sec // 60
#   sec = sec % 60
#   mins = mins % 60
#   print("Time Lapsed = {0}:{1}".format(int(mins),int(sec)))

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

input('Press ENTER to start the game')

p=[]
q=0
while(q<50):
  r=random.randint(0,49)
  if r not in p:
    p.append(r)
    q+=1

df = pd.read_excel('/Hackathon_statement 2.xlsx')
sent=df['English sentences']

ch=1
while ch==1:
  cc+=1
  sentence=sent[p[cc-1]]
  tokens = nltk.word_tokenize(sentence)
  pos_tag = nltk.pos_tag(tokens)
  pos_tags=[]
  for i in pos_tag:
    pos_tags.append(list(i))

  pos_tags=pos_tags[:-1]

  for i in pos_tags:
    if i[1] in noun:
      i[1]='noun'
    elif i[1] in pronoun:
      i[1]='pronoun'
    elif i[1] in verb:
      i[1]='verb'
    elif i[1] in adverb:
      i[1]='adverb'
    elif i[1] in adjective:
      i[1]='adjective'
    elif i[1] in preposition:
      i[1]='preposition'
    elif i[1] in conjunction:
      i[1]='conjunction'
    elif i[1] in interjuction:
      i[1]='interjuction'
    elif i[1] in article:
      i[1]='article'

  x=[]
  for i in pos_tags:
    x.append(i[1])
  x=list(set(x))

  random.shuffle(x)

  y=[]
  for i in pos_tags:
    y.append(i[0])

  start_time = time.time()
  print('\nQuestion '+str(cc))
  print(sentence+'\n\nOptions')
  for i in range(len(x)):
    print(str(i+1)+'. '+x[i])
  print()
  z=[]
  for i in range(len(y)):
    print(y[i] +' ',end=' ')
    z.append(input())
  end_time = time.time()
  time_lapsed = end_time - start_time

  w=[]
  for i in pos_tags:
    w.append(i[1])

  c,d=0,0
  for i in range(len(w)):
    if (w[i])==x[int(z[i])-1]:
      c+=1
    else:
      d+=1
  u.append(c)
  v.append(d)
  cp=(c/(c+d))*100
  dp=(d/(c+d))*100
  print("\nCorrect: "+str(cp)+"%")
  print("Wrong: "+str(dp)+"%")
  print('Time Taken: '+str(int(time_lapsed))+'sec')
  sc=0
  pas=False
  if cp>35:
    pas=True
  if int(time_lapsed)<30 and pas:
    sc=3
  elif int(time_lapsed)<45 and pas:
    sc=2
  elif int(time_lapsed)<60 and pas:
    sc=1
  score+=sc
  print('Score: '+str(sc)+'pts\n\n')
  ti.append(int(time_lapsed))
  print('Choices\n1.Next Question\n2.Quit Game')
  ch=int(input('Enter Choice: '))

sum=0
for i in ti:
  sum+=i

avg=sum/cc

print('\n\nReport\n------')
print('Total time taken: '+str(sum)+'sec')
print('Average time for each question: '+str(avg)+'sec')

uu,vv=0,0
for i in u:
  uu+=i
for i in v:
  vv+=i
print("Total Correct: "+str((uu/(uu+vv))*100)+'%')
print("Total Wrong: "+str((vv/(uu+vv))*100)+'%\n\n\n')

print('Total Score: '+str(score)+'pts/'+str(cc*3)+'pts')
res=(score/(cc*3))*100
if res<35:
  print('Result: Fail')
else:
  print('Result: Pass')

# flag=0
# for i in range(len(u)):
#   if not(((u[i]/(u[i]+v[i]))*100)>35):
#     flag=1
# if flag==1:
#   print('Result: Fail')
# else:
#   print('Result: Pass')

```
